### my binder test custom env

This environment was created to test with mybinder and github.
During initial creation, the SQL magic formatting from Ipython-sql had trouble installing and wouldn't connect to the AWS instance.

This environment had a successful connection after the following:

The Ipython-sql env was loaded before the Jupyter notebook env. Their seemed to be a conflict when Jupyter was installed first.

After that '%load_ext sql' or 'load_ext sql' started working, but the AWS instance could not connect because Psycopg2 was not properly loaded in spite of it being listed in the installed environments.

The link below provided the step necessary to connect AWS instance: 
#https://jakevdp.github.io/blog/2017/12/05/installing-python-packages-from-jupyter/

'import sys' and '!{sys.executable} -m pip install psycopg2'

were used to install Pyscopg2 successfull and the AWS instance could connect.
